```python
#Importar librerias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Cargar los datos presentes en la hoja de balance calculada. 
df=pd.read_csv("/home/jovyan/Tesis/HojaBalanceF.csv")
df
```

```python
#Generar una variable por cada columna de la base de datos para su posterior utilizaci칩n. 
anios=df["A침o"]
Q1=df["Flujo de efectivo Q1 ($/MNX)"]
Q2=df["Flujo de efectivo Q2 ($/MNX)"]
Q3=df["Flujo de efectivo Q3 ($/MNX)"]
```

```python
#Crear la figura. 
fig, ax = plt.subplots(figsize=(10, 5))
 
#Plasmar los valores presentes en las variables delcaradas anteriormente en l칤neas. 
plt.plot(anios, Q1, color='#FF1700', linestyle='-', alpha=0.9, label="Q1")  
plt.plot(anios, Q2, color='#FBFF00', linestyle='-', alpha=0.9,label="Q2")  
plt.plot(anios, Q3, color='#337357',linestyle='-', alpha=0.9,label="Q3")  

#Configurar la vista del gr치fico.
plt.title("Flujos de efectivo con la implementaci칩n de los cuantiles de producci칩n el칠ctrica", fontsize = 14, fontweight = "bold")
plt.ylabel(r'$\$MXN$');
plt.ylim([-4000000,10000000])
plt.xlabel("A침o")
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.tight_layout()
plt.axhline(y=0, color='#C7C8CC', linestyle='--')
plt.savefig("FlujosEfect.svg", format="svg",dpi=300, bbox_inches='tight')
plt.show()
```
