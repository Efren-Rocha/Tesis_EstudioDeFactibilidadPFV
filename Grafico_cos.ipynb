```python
#Importar librerias.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Cargar hoja de cálculo.
df=pd.read_csv("/home/jovyan/Tesis/TethaF.csv")
df
```

```python
#Limpiar los datos: convertir la columna "h" en índice.
df.index=df.h
df=df.drop('h',axis=1)
```

```python
#Seleccionar únicamente la columna cos (0) formula 2. 
column=df["cos(θ) formula 2"]
```

```python
#Generar un nuevo df con las columnas que nos interesan.
df2= {"Días": np.arange(1,365),"cos":column}
df3 = pd.DataFrame(df2)
```

```python
#Generar el df definitivo con el índice de 1 a 365 y con la columna seleccionada. 
df3.index=df3.Días
df3=df3.drop('Días',axis=1)
```

```python
#Generar una variable que contenga los 365 días del año para su representación como eje "X". 
dias = np.arange(1, 365)

# Crear la figura.
plt.figure(figsize=(12, 6))

#Configurar la vista del gráfico. 
plt.plot(dias, df3["cos"], label=r"$Cos(θ)$", color="#EB5A3C")
plt.title("Variación del Cos(θ) a lo largo del año", fontsize=14)
plt.xlabel("Días", fontsize=12)
plt.ylabel(r"$Cos(θ)$", fontsize=12)
plt.tight_layout()

plt.savefig("Var_cosTheta.svg", format="svg")
plt.show()
```
